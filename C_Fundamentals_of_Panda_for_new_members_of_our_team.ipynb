```python
#make a dataframe using arrays
#basically what happens in dataframe is that every row and column will get a key name(index and column) that is a string
arr=np.array([[1,2,3,4],[0,0,0,0],[9,9,9,9]])
arrdf=pd.DataFrame(arr)
#col and index have to be string
col=['st1','st2','st3','st3']
arrdf.columns=col
idx=['c1','c2','c3']
arrdf.index=idx
#if you wanna update index you change the col or idx lists
arrdf
```

```python
#Notice the number of brackets
arrdf.index[0:3] #or arrdf.index[[0,3]]
```

```python
# query a row or column with iloc (index position)
arrdf.iloc[:2,:]
```

```python
# query a row with loc (index names)
arrdf.loc['c1']['st1':'st2']
```

```python
# query a row with a name of one axis and the numbers in the other axis
arrdf.loc['c1'][0:2]
```

```python
# query a row with a names of one axis and the numbers in the other axis
arrdf.loc['c1':'c3'][0:2]
```

```python
# query a column with a names of one axis and the numbers in the other axis
arrdft=arrdf.T
arrdft.loc['st1':'st3'][0:2]
```

```python
#remove a column using column label
arrdf=arrdf.drop(['st1','st2'],axis=1)
arrdf
```

```python
#remove a column using column number
arrdf=arrdf.drop(arrdf.columns[[0,1]],1)
arrdf
```

```python
#add a column
arrdf.loc[:]['stdx']=[8,8,8]
arrdf
```

```python
#add a row
#arrdf.loc['c4']=[0,0,0,0,None]   
arrdf.iloc[3,:]=[0,0,0,0,None]
arrdf
```

```python
#concat two data frames or a dataframe and a series

arrdf=pd.DataFrame([[1,2,3,4],[0,0,0,0],[9,9,9,9]])

new_df=pd.Series([7,7,7])
arrdf=pd.concat([new_df,arrdf],axis=1)
arrdf
```

```python
#concat two data frames or a dataframe and a series
arrdf=pd.DataFrame([[1,2,3,4],[0,0,0,0],[9,9,9,9]])

new_df=pd.DataFrame([[7,7,7,7]])
arrdf=pd.concat([new_df,arrdf],axis=0)
indx=['0','1','2','3']
arrdf.index=indx
arrdf
```

```python
#example dataframe with a dictionary and addingincreasing index using a for
dic1={'Mo': [2,18,12,15],
     'Alex':[0,2,5,7],
     'Narges':[20,18,19,20]
      }
dic2=pd.DataFrame(dic1,index=['course' + str(n) for n in range(len(dic1['Mo']))])
pd.DataFrame(dic2)
```

```python
dic2=[{'name':'Alex','class':'Chemistry','score':[20,12]},
      {'name':'Jack','class':'Math','score':[19,20]},
      {'name':'Jennifer','class':'Physics','score':[14,15]}]

#put a list of dictionaries into a dataframe
dic2pd=pd.DataFrame(dic2,index=['student'+ str(n) for n in range(len(dic2))])
dic2pd
```

```python
# add a column by name
dic2pd['lastname']=None
print(dic2pd,'\n')
```

```python
#practical example of a comma separated file, comma separated file (cvs)==== should execute the cells in order
#!cat SPY_3.csv
dt=pd.read_csv('SPY_3.csv')
dt.head(3)
```

```python
# get rid of spaces in the column names, we should do this begore hand as some names may include a space -Importanttt
dt=dt.rename(mapper=str.strip, axis='columns')
```

```python
dt=dt.set_index('Date')
dt.head()
```

```python
# columns shown
dt.columns
```

```python
dt=dt.rename(columns={'Open':'newopen','High':'newhigh', 'Date':'newdate'})
dt.head(3)
```

```python
dt=dt.rename(index={"2020-12-31":2})
dt.head(3)
```

```python
# lower case columns
dtcol=list(dt.columns)
dtcol=[x.lower().strip() for x in dtcol]
dt.columns=dtcol
dt.head(3)
```
