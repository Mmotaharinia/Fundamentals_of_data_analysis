```python
#Matrix elements

eye1=np.eye(7)
for j in range(len(eye1)):
  for i in range(len(eye1[0])):
    if eye1[j,i]>0:
      eye1[j,i]=1000
    elif eye1[j,i]==0:
      eye1[j,i]=2
    i=i+1
  j=j+1

print('eye1=','\n',eye1)
```

```python
dic2=[{'name':'Alex','class':'Chemistry','score':20},
      {'name':'Jack','class':'Math','score':19},
      {'name':'Jennifer','class':'Physics','score':14}]


dic2pd=pd.DataFrame(dic2,index=['student'+ str(n) for n in range(len(dic2))])
# query the dictionary.
print('dic2pd=','\n',dic2pd,'\n','\n')
print(dic2[1]['class'],'\n','\n')

#in most cases we need to directly work with the datafarame

#query a row in a dataframe
print(dic2pd.loc['student0'],'\n','\n')
print(dic2pd.loc['student0']['name'],'\n')

#query a column in a dataframe
dic3pd=dic2pd.T
print(dic3pd.loc['name'],'\n')

#query multiple rows together
print(dic2pd.loc[:'student1',['name','score']])
```

```python
#Add or remove a row/column to a dataframe
dic2=[{'name':'Alex','class':'Chemistry','score':20},
      {'name':'Jack','class':'Math','score':19},
      {'name':'Jennifer','class':'Physics','score':14}]

dic2pd=pd.DataFrame(dic2,index=['student'+ str(n) for n in range(len(dic2))])
print(dic2pd,'\n')
# add a column
dic2pd['lastname']=None
print(dic2pd,'\n')
#remove a column
del dic2pd['lastname']
print(dic2pd,'\n')
```
